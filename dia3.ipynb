# crear una funcion que reciba dos parametros (una palabra y un numero) y que retorne True si el numero es igual a la cantidad de las letras de la palabra. en caso contrario, que retorne False.
# pista: cantidad de letras de una palabra en python

1 - crear una funcion que permita recibir 2 parametros (palabra, numero)
2- definir la funcion  verificar(palabra, numero)
3- ingresar la palabra y numero con input()
4- verificar cantidad de letras con len 
5- si len(palabra) == numero:
6- return = true
7- else:
8- return = false
9- llamar a la funcion verificar()

```python
palabra = input('ingrese una palabra')
numero = int(input('ingrese un numero'))

def verificar(p1, n1):
    if len(p1) == n1:
        return True
    else:
        return False

verificar(palabra, numero)
```
