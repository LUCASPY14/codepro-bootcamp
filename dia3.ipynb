```python
temperatura = int(input('ingrese la temperatura actual:'))

def clima(temperatura):
    if temperatura >= -10 and temperatura <=18:
        print('fresquete')
    elif temperatura >= 19 and temperatura <= 30:
        print("Calor'i")
    elif temperatura >= 31 and temperatura <= 45:
        print('Hakuuuuuu')
    
    else:
        print("Namano mba'e!!")

clima(temperatura)
```
